### 1.Create a function to define if a string is a palindrome or not. (A palindrome is a word or sentence that has the same letters backwards and forwards)

### a) Replacing whitespaces is a good way to start b) Watch out for upper/lower case differences

```python
def palindrome(string):
    
    if(string.replace(" ","").lower()==string[::-1].replace(" ","").lower()):
        print("The string is a palindrome")
    else:
        print("The string isn't a palindrome")
```

```python
palindrome('palindrome')
```

```python
### Try with words/sentences like:

###"Was it a car or a cat I saw"
###"Ceci n'est pas un palindrome"
###GCCTTCCG
###CGTAATGC
###Anita lava la tina
###kayak
```
