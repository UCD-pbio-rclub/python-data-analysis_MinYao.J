### Please use matplotlib to draw a 3D random walk for 3000 steps.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Generate the randam walk data.

```python
N = 3000
R = (np.random.rand(N)*6).astype("int") #Random values in a given shape.
x = np.zeros(N)
y = np.zeros(N)
z = np.zeros(N)
x[ R==0 ] = -1; x[ R==1 ] = 1
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x)
y = np.cumsum(y)
z = np.cumsum(z)
```

### Plot with same color.

```python
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.6)
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
```

### Plot with changing color

```python
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
cm = plt.get_cmap('jet') # Get a colormap instance
ax.set_prop_cycle('color',[cm(1.*i/(x.shape[-1]-1)) for i in range(x.shape[-1]-1)]) 
# Set the property cycle of the Axes. The property cycle controls the style properties such as color.
# get the current shape of an array
for i in range(x.shape[-1]-1):
    ax.plot([x[i+1],x[i]], [y[i+1],y[i]], [z[i+1],z[i]],alpha=0.6)
ax.scatter(x[-1],y[-1],z[-1],facecolor=cm(1))
plt.savefig('3d_random_walk_static.png', bbox_inches='tight', pad_inches=0.02, dpi=300)
plt.show()
```

### 1. Create a smiley face using 4 subplots

```python
import numpy as np
import pandas as pd
PREVIOUS_MAX_ROWS = pd.options.display.max_rows
pd.options.display.max_rows = 20
np.random.seed(12345)
import matplotlib.pyplot as plt
import matplotlib
plt.rc('figure', figsize=(10, 6))
np.set_printoptions(precision=4, suppress=True)
from matplotlib.patches import Circle, Wedge, Polygon
```

```python
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0, hspace=0)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

circ1 = plt.Circle((0.5, 0.5), 0.15, color='k', alpha=0.9)
circ12 = plt.Circle((0.5, 0.5), 0.05, color='k', alpha=1)
circ2 = plt.Circle((0.5, 0.5), 0.15, color='k', alpha=0.9)
circ22 = plt.Circle((0.5, 0.5), 0.05, color='k', alpha=1)
sector1 = Wedge((1, 0.7), 0.6, -180, -90, color='r')
sector2 = Wedge((0, 0.7), 0.6, -90, 0, color='r')
ax1.add_patch(circ1)
ax1.add_patch(circ12)
ax2.add_patch(circ2)
ax2.add_patch(circ22)
ax3.add_patch(sector1)
ax4.add_patch(sector2)
```
