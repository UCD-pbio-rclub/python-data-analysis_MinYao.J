## Q11 Changing topics...generate and plot a 2D random walk (see Chapter 4.7)

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# 1D

```python
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
plt.plot(walk[:1000])
```

# 2D

## pylab

```python
import pylab 
  
# defining the number of steps 
n = 1000
  
#creating two array for containing x and y coordinate 
#of size equals to the number of size and filled up with 0's 
x = np.zeros(n) 
y = np.zeros(n) 
  
# filling the coordinates with random variables 
for i in range(1, n): 
    val = random.randint(1, 4) 
    if val == 1: 
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif val == 2: 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif val == 3: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
      
# plotting stuff
pylab.title("Random Walk ($n = " + str(n) + "$ steps)") 
pylab.plot(x, y) 
pylab.savefig("rand_walk"+str(n)+".png",bbox_inches="tight",dpi=600) 
pylab.show()
```

## Other methods

```python
def xywalk(n):
    x = np.zeros((n))
    y = np.zeros((n))
    for i in range(0,n):
        walk = random.random()
        if 0 < walk < .25:
            x[i] = x[i-1] + 1
        elif .25 < walk < .5:
            x[i] = x[i-1] - 1
        elif .5 < walk < 0.75:
            y[i] = y[i-1] + 1
        else:
            y[i] = y[i-1] - 1
    return x, y



x, y = xywalk(1000)

# create a figure
fig = plt.figure()
# create a plot into the figure
ax = fig.add_subplot(111)
# plot the data
ax.plot(x,y)
```
